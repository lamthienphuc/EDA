```python
tn_movie = pd.read_csv('tn.movie_budgets.csv')
tn_movie.head()
```

```python
# basic stripping 
tn_movie['production_budget'] = tn_movie['production_budget'].apply(lambda x: x.lstrip('$'))
tn_movie['domestic_gross'] = tn_movie['domestic_gross'].apply(lambda x: x.lstrip('$'))
tn_movie['worldwide_gross'] = tn_movie['worldwide_gross'].apply(lambda x: x.lstrip('$'))
```

```python
# convert the values to float numbers
tn_movie['production_budget'] = tn_movie['production_budget'].str.replace(',', '').astype(float)
tn_movie['domestic_gross'] = tn_movie['domestic_gross'].str.replace(',', '').astype(float)
tn_movie['worldwide_gross'] = tn_movie['worldwide_gross'].str.replace(',', '').astype(float)
tn_movie.head()
```

```python
# set the index to the id
tn_movie.set_index('id', inplace = True)
```

```python
# create a new column that calculates foreign gross.
tn_movie['foreign_gross'] = tn_movie['worldwide_gross'] - tn_movie['domestic_gross']
tn_movie.head()
```

```python
# taking a look at the biggest production budgets
# all values are in original form, they would be reduced to millions eventually
tn_movie.sort_values('production_budget', ascending = False)
```
